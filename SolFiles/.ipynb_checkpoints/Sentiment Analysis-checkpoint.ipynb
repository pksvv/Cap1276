## Data Preprocessing
Lower the text in comments

```python
comment_data['comments'].head()
```

```python
import nltk
nltk.download("stopwords")
```

```python
nltk.download("punkt")
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
import string
import re
import nltk
```

```python
stopwords.words('english')
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(line):
    word_tokens = word_tokenize(line)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    return " ".join(filtered_sentence)
```

```python
comment_data['stop_comments'] = comment_data['comments'].apply(lambda x : remove_stopwords(x))
reply_data['stop_reply'] = reply_data['reply'].apply(lambda x : remove_stopwords(x))
```

```python
comment_data.head()
```

```python
tokenized_comment = comment_data['stop_comments'].apply(lambda x: x.split())
tokenized_reply = reply_data['stop_reply'].apply(lambda x: x.split())
```

```python
tokenized_comment.head()
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
tokenized_comment = tokenized_comment.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_reply = tokenized_reply.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_comment.head()
```

```python
for i in range(len(tokenized_comment)):
    tokenized_comment[i] = ' '.join(tokenized_comment[i])
```

```python
for i in range(len(tokenized_reply)):
    tokenized_reply[i] = ' '.join(tokenized_reply[i])
```

```python
tokenized_comment
```

```python
comment_data['token_comments'] = tokenized_comment
reply_data['token_reply'] = tokenized_reply
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(comment_data['token_comments'],comment_data['comment_pol_cat'],test_size = 0.2,random_state = 324)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(stop_words='english')
# bag-of-words feature matrix
bow_train = bow_vectorizer.fit_transform(X_train)
bow_test = bow_vectorizer.transform(X_test)
```

```python
bow_vectorizer.vocabulary_
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(bow_train,y_train)
```

```python
lr.score(bow_train,y_train)
```

```python
lr.score(bow_test,y_test)
```

```python
expected = y_test
predicted = lr.predict(bow_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(expected,predicted)
```

```python
print(classification_report(expected, predicted))
```
