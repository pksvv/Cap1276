## Remove Emojis
In this step we will remove emojis from the comments using demoji

```python
## this step is necessary as it will download the latest emojis if updated so as to remove all from the given texts

#demoji.download_codes()
```

```python
comment_df['clean_comments'] = comment_df['comment'].apply(lambda x: demoji.replace(x,""))
reply_df['clean_reply'] = reply_df['reply'].apply(lambda x: demoji.replace(x,""))
```

Now add new feature language inthe dataframe which signify the language of particular comment

```python
comment_df['comment_language'] = 0
reply_df['reply_language'] = 0
```

```python
for i in range(0,len(comment_df)):
    temp = comment_df['clean_comments'].iloc[i]
    try:
        comment_df['comment_language'].iloc[i] = detect(temp)
    except:
        comment_df['comment_language'].iloc[i] = "error"
```

```python
for i in range(0,len(reply_df)):
    temp = reply_df['clean_reply'].iloc[i]
    try:
        reply_df['reply_language'].iloc[i] = detect(temp)
    except:
        reply_df['reply_language'].iloc[i] = "error"
```

```python
print(reply_df['reply_language'].value_counts())
print(comment_df['comment_language'].value_counts())
```

```python
english_comm = comment_df[comment_df['comment_language'] == 'en']
english_reply = reply_df[reply_df['reply_language'] == 'en']
```

## Remove brackets & special characters using Regular Expressions
Now first write the grammar to remove special characters from the comments

```python
regex = r"[^0-9A-Za-z'\t]"
```

```python
comments_data = en_comments.copy()
reply_data = en_reply.copy()
```

```python
# Comment remove secial char using regex
comments_data['reg'] = comments_data['clean_comments'].apply(lambda x:re.findall(regex,x))
comments_data['regular_comments'] = comments_data['clean_comments'].apply(lambda x:re.sub(regex,"  ",x))
# reply remove secial char using regex
reply_data['reg'] = reply_data['clean_reply'].apply(lambda x:re.findall(regex,x))
reply_data['regular_reply'] = reply_data['clean_reply'].apply(lambda x:re.sub(regex,"  ",x))
```

```python
comments_data.head()
```

```python
reply_data.head()
```

```python
comment_dataset = comments_data[['commentId','videoId','commentLikesCount','regular_comments']].copy()
reply_dataset = reply_data[['replyId','videoId','parentCommentId','replylikesCount','regular_reply']].copy()
```

```python
comment_dataset = comment_dataset.rename(columns = {"regular_comments":"comments"})
reply_dataset = reply_dataset.rename(columns = {"regular_reply":"reply"})
```

```python
comment_dataset.to_csv("comment_clean_dataset.csv",index = False)
reply_dataset.to_csv("reply_clean_dataset.csv",index = False)
```
